# 2. Show Bar Graphs representing the Departure Time & Arrival Time

```python
df["departure_time"].value_counts()
```

```python
df["arrival_time"].value_counts()
```

```python
plt.figure(figsize = (16,4))
plt.subplot(1,2,1)

plt.bar( df['departure_time'].value_counts().index , df['departure_time'].value_counts().values, color = ["g"] )
plt.title("Departure Time")
plt.xlabel("D. Time")
plt.ylabel("D. Freq")

plt.subplot(1,2,2)

plt.bar( df['arrival_time'].value_counts().index, df['arrival_time'].value_counts().values, color = ["y"])
plt.title("Arrival Time")
plt.xlabel("A. Time")
plt.ylabel("A. Freq")

plt.show()
```

# 3. Show Bar Graphs representing the Source City & Destination City

```python
df["source_city"].value_counts()
```

```python
df["destination_city"].value_counts()
```

```python
plt.figure( figsize= (16,4))

plt.subplot(1,2,1)

plt.barh( df['source_city'].value_counts().index , df['source_city'].value_counts().values, color = ["r"])
plt.title("Source Cities with No. of flights")
plt.ylabel("Cities")
plt.xlabel("No. of flights")

plt.subplot(1,2,2)

plt.barh( df['destination_city'].value_counts().index , df['destination_city'].value_counts().values, color = ["b"])
plt.title("Destination Cities with No. of flights")
plt.ylabel("Cities")
plt.xlabel("No. of flights")

plt.show()
```

# 5. Does ticket price change based on the departure time and arrival time?

```python
df.groupby("departure_time")["price"].mean()
```

```python
df.groupby("arrival_time")["price"].mean()
```

```python
sns.catplot(x="departure_time",y="price",kind="bar",data=df)
plt.figure(figsize=(6,15))
plt.show()
```

```python
sns.catplot(x="arrival_time",y="price",kind="bar",data=df)
plt.figure(figsize=(6,15))
plt.show()
```
